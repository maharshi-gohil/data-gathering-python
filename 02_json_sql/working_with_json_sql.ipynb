# Working with JSON and SQL Data in Python

This notebook demonstrates how to load data from JSON files, JSON APIs, and SQL databases into pandas DataFrames for analysis.

It covers:
- Reading local JSON files
- Reading JSON data from URLs
- Connecting Python with MySQL
- Executing SQL queries
- Converting SQL query results into pandas DataFrames

````python
import pandas as pd
````

## Reading JSON Data

JSON stands for **JavaScript Object Notation**.

It is a widely used data format for storing and exchanging structured data.

JSON is commonly used in:
- API responses
- configuration files
- web applications
- data exchange between systems

````python
df = pd.read_json('train.json')
df.head()
````

````python
df.shape
````

````python
df.info()
````

## Read JSON Data from a URL

JSON data can also be loaded directly from a URL.

This is useful when working with APIs or publicly available JSON endpoints.

````python
url = 'https://api.exchangerate-api.com/v4/latest/INR'

df = pd.read_json(url)
df.head()
````

````python
df.shape
````

## Working with SQL Data

SQL stands for **Structured Query Language**.

It is used to communicate with relational databases.

SQL can be used to:
- create databases
- create tables
- insert records
- filter records
- retrieve data
- join multiple tables

## Role of XAMPP in This Notebook

In this workflow:

- **Python** connects to the database
- **pandas** converts SQL query results into DataFrames

This setup is useful for practicing database workflows locally.

## Install MySQL Connector

To connect Python with MySQL, install the `mysql-connector-python` library.

This library acts as a bridge between Python and MySQL.

````python
import mysql.connector
````

## Create MySQL Database Connection

To connect Python with MySQL, the following details are required:

- host name
- username
- password
- database name

For a default XAMPP setup:

- host: `localhost`
- user: `root`
- password: empty string

````python
connection = mysql.connector.connect(host='localhost',user='root',password='',database='world')
connection
````

## Read SQL Table into pandas DataFrame

pandas provides `pd.read_sql_query()` to execute SQL queries and load the result into a DataFrame.

It requires:

1. SQL query
2. Database connection object

````python
df = pd.read_sql_query("SELECT * FROM countrylanguage", connection)
df.head()
````
