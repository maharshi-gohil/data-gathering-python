# Working with CSV Files using Pandas

In this notebook, we will learn how to work with CSV files using Pandas.

CSV files are one of the most commonly used file formats in Machine Learning and Data Science.

We will cover:
- Reading CSV files
- Loading CSV from URL
- Handling separators
- Working with headers
- Selecting columns
- Handling missing values
- Parsing dates
- Reading large files in chunks

```python
import pandas as pd
```

# Reading a CSV File

`pd.read_csv()` is used to load CSV files into a Pandas DataFrame.

```python
df = pd.read_csv('aug_train.csv')
df.head()
```

```python
df.shape
```

# Reading CSV from URL

Sometimes datasets are stored online.

We can fetch the data using the `requests` library and then load it into Pandas.

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
df.head()
```

```python
df.shape
```

# Working with Different Separators and Adding Column Names

By default, Pandas expects comma-separated values.

Some files use:
- tabs
- semicolons
- pipes

In such cases, we use the `sep` parameter.

Sometimes datasets do not contain column names.

We can manually provide them using the `names` parameter.

```python
df = pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
df.head()
```

# Setting an Index Column

If a dataset already contains a unique ID column, we can use it as the index.

```python
df = pd.read_csv('aug_train.csv', index_col = 'enrollee_id')
df.head()
```

# Header Parameter

Sometimes the actual column names are not present in the first row.

In such cases, we use the `header` parameter.

```python
df = pd.read_csv('test.csv', header = 1)
df.head()
```

# Selecting Specific Columns

Real-world datasets may contain many unnecessary columns.

Using `usecols` helps load only required columns.

```python
df = pd.read_csv('aug_train.csv', usecols = ['enrollee_id','gender','education_level'])
df.head()
```

# Skipping Rows

Sometimes datasets contain unnecessary rows.

We can skip them using `skiprows`.

```python
df = pd.read_csv('aug_train.csv', skiprows = [0,2])
df.shape
```

# Loading Limited Rows

Large datasets can consume a lot of memory.

Using `nrows` helps load only a fixed number of rows.

```python
df = pd.read_csv('aug_train.csv', nrows=100)
df
```

# Handling Encoding Issues

Some datasets use different text encodings.

In such cases, we use the `encoding` parameter.

```python
df = pd.read_csv('zomato.csv',encoding='latin-1')
df.head()
```

# Handling Bad Lines and Dtype Warnings

Some real-world CSV files may contain:
- corrupted rows
- extra separators
- mixed datatypes

In this dataset:
- some rows contain extra columns
- some columns contain mixed datatypes

To handle this:

- `on_bad_lines="skip"` skips corrupted rows
- `low_memory=False` helps Pandas properly detect datatypes
- `encoding="latin-1"` handles special characters

```python
columns = [
    "ISBN",
    "Book-Title",
    "Book-Author",
    "Year-Of-Publication",
    "Publisher",
    "Image-URL-S",
    "Image-URL-M",
    "Image-URL-L"
]
df = pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1", on_bad_lines="skip", low_memory=False, names=columns)
df.head()
```

# Understanding Data Types

Pandas automatically detects data types.

We can manually define them using `dtype`.

```python
df = pd.read_csv("aug_train.csv", dtype={"target": "int"})
df.info()
```

# Parsing Date Columns

Dates are usually loaded as strings.

Using `parse_dates` converts them into datetime format.

```python
df = pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

# Using Converters

Converters allow us to transform values while loading data.

```python
def rename_team(team):
    
    if team == "Royal Challengers Bangalore":
        return "RCB"
    
    return team
```

```python
rename_team("Royal Challengers Bangalore")
```

```python
df = pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename_team})
df.head()
```

# Handling Missing Values

Some datasets contain custom missing value placeholders like:
- "-"
- "missing"
- "not available"

We can convert them into NaN values using `na_values`.

```python
df = pd.read_csv('aug_train.csv', na_values=['Male',])
df.head()
```

# Reading Large Datasets in Chunks

Large datasets may not fit completely into memory.

Using `chunksize` helps process the data in smaller parts.

```python
dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
for chunk in dfs:
    print(chunk.shape)
```
