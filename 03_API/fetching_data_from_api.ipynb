# Fetching Data from API using Python

In this notebook, we will fetch movie data from TMDB API, convert JSON data into a Pandas DataFrame, and export the final dataset as a CSV file.

We will learn:

- What an API response looks like
- How to use `requests`
- How to convert JSON into DataFrame
- How to handle pagination
- How to save data as CSV

```python
import requests
import pandas as pd
from dotenv import load_dotenv
import os
```

## API Key Safety

We should never write our API key directly inside the notebook because this notebook will be uploaded to GitHub.

Instead, we will store the API key inside a `.env` file.

```python
load_dotenv()

api_key = os.getenv('TMDB_API_KEY')

if api_key:
    print("API key loaded successfully")
else:
    print("API key not found")
```

## Fetch Data from TMDB API

We will use the Top Rated Movies endpoint:

`https://api.themoviedb.org/3/movie/top_rated`

This endpoint returns movie data in JSON format.

```python
url = f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&page=1'

response = requests.get(url)
print(response.status_code)
```

```python
data = response.json()
data
```

```python
data.keys()
```

```python
movies = data['results']
movies[0]
```

```python
df = pd.DataFrame(movies)
df.head()
```

```python
df = df[
    [
        "id",
        "title",
        "overview",
        "release_date",
        "popularity",
        "vote_average",
        "vote_count"
    ]
]

df.head()
```

## Pagination

The API gives data page by page.

One page contains around 20 movies.  
To collect more movies, we need to loop through multiple pages.

```python
total_pages = data['total_pages']
total_results = data['total_results']

print("Total pages:", total_pages)
print("Total results:", total_results)
```

```python
all_movies=[]

for page in range(1,11):
    url = f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&page={page}'
    
    response = requests.get(url)
    data = response.json()
    movies = data['results']
    all_movies.extend(movies)

len(all_movies)
```

```python
movies_df = pd.DataFrame(all_movies)
movies_df.head()
```

```python
movies_df = movies_df[
    [
        "id",
        "title",
        "overview",
        "release_date",
        "popularity",
        "vote_average",
        "vote_count"
    ]
]

movies_df.head()
```

```python
movies_df.shape
```

```python
movies_df.to_csv('TMDB_top_rated_movies.csv', index=False)
print("csv created successfully")
```
